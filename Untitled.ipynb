```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV
jnj_data = pd.read_csv('results/predictions/xgboost_predictions_JNJ.csv', parse_dates=['Date'])

# Plot actual vs predicted trend for JNJ
plt.figure(figsize=(10,5))
plt.plot(jnj_data['Date'], jnj_data['Actual'], label='Actual', color='blue')
plt.plot(jnj_data['Date'], jnj_data['Predicted'], label='XGBoost Predicted', color='red')
plt.title('JNJ Actual vs XGBoost Predicted Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()

  
```
